```python
import pandas as pd
import matplotlib.pyplot as plt
import io
import requests
import seaborn as sns

# Configuraciones opcionales para que los gráficos tomado de lo estudiado en clase 4
%matplotlib inline
sns.set(style="whitegrid", palette="deep")
```

```python
# 1. Tomar archivo CSV base "Hipertensión Arterial Mexico" desde un raw en GitHub de mi repositorio público
url = "https://raw.githubusercontent.com/rgroesman/74915-data_science_fundamentos/refs/heads/main/Hipertension_Arterial_Mexico.csv"
response = requests.get(url)
content = response.content

# 2. Cargar el archivo CSV en un DataFrame (de Panda)
#df = pd.read_csv(url) # Modo común sin intermediarios de ningún tipo, el resultado NO ES EL ESPERADO
#df = pd.read_csv(io.StringIO(content.decode('utf-8'))) # El delimitador de columnas no es el estandar COMO, es PUNTO Y COMA
df = pd.read_csv(io.StringIO(content.decode('utf-8')), sep=';')

# 3. Muestra el DataFrame por consola
#print("Data Frame por consola (A MODO DE PRUEBA):")
#print(df)

# 4. Muestras primeros registros del data frame
df.head()
```

## Cálculo de medidas básicas tomando campo ejemplo Edad


```python
# 5. Calcula la media y la mediana del campo Edad
print("\nEstadísticas:\n")
print("Media de la edad:",df["edad"].mean())
print("Mediana de la edad:",df["edad"].median())

```

##Gráfico en base a la Edad

```python
# Histograma de la variable Edad
plt.figure(figsize=(8,4))
sns.histplot(df['edad'], kde=True, color='blue')
plt.title("Distribución por Edad")
plt.xlabel("Edad")
plt.ylabel("Cant. de casos análizados")
plt.show()

# Estadísticas descriptivas adicionales
print("Media:", df['edad'].mean())
print("Mediana:", df['edad'].median())
print("Desv. Estándar:", df['edad'].std())
```
